Recessions are periods of economic contraction having a significant impact on various industries. Typically, a recession is characterized by a significant decline in the gross domestic product over a time period, leading to widespread unemployment, loss of income, and reduced business activity.

Here is a dataset of the monthly GDP of the United Kingdom from 2020 to 2022. Below are all the features in the dataset:

Time Period: Monthly time period
GDP Growth: The growth rate of GDP every month

```python
import pandas as pd
import plotly.graph_objs as go
import plotly.express as px
import plotly.io as pio
pio.templates.default = "plotly_white"
import matplotlib.pyplot as plt
```

#### GDP growth over time

```python
ra
```

```python
ra['Time Period'] = pd.to_datetime(ra['Time Period'])

# Sort the DataFrame by 'Date' column
ra.sort_values(by='Time Period', inplace=True)

# Calculate GDP growth rate over time
ra['GDP_Growth'] = ra['GDP Growth'].pct_change() * 100

# Print the DataFrame to see the results
print(ra)
```

there  will be calculate the GDP growth rate as a percentage change from the previous period and store it in a new column named 'GDP_Growth'. 

```python
ra['Date'] = pd.to_datetime(ra['Date'])

# Sort the DataFrame by 'Date' column
ra.sort_values(by='Date', inplace=True)

# Group by quarter and calculate the sum of GDP for each quarter
ra_quarterly = ra.groupby(pd.Grouper(key='Date', freq='Q')).sum()

# Print the quarterly DataFrame
print(ra_quarterly)
```

```python
ra['Time Period'] = pd.to_datetime(ra['Time Period'], format='/%m/%Y')
ra.set_index('Time Period', inplace=True)
quarterly_data = ra.resample('Q').mean()
print(quarterly_data.head())
```
